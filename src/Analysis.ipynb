```python
df = pd.read_csv('../data/pairwise_travel_times.csv')
df.columns = df.columns.str.strip() # the column names have endspaces etc.
df['duration_walking_min'] = np.ceil(df['duration_seconds'] / 60).astype(int) # Convert to minutes and round up
df.rename(columns={'duration_seconds': 'duration_walking_sec'}, inplace=True)
df.drop(columns=['mode', 'distance_text', 'duration_text', 'duration_in_traffic_seconds', 'duration_in_traffic_text'], inplace=True)
df['origin'] = df['origin'].astype('string')
df['destination'] = df['destination'].astype('string')
df['duration_public_transport_sec'] = np.nan
df['duration_public_transport_min'] = np.nan
df.head()
```
